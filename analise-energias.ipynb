```python
import pandas as pd

# Carregar o arquivo
caminho_arquivo = 'siga-empreendimentos-geracao.csv'
df = pd.read_csv(caminho_arquivo, sep=';', encoding='latin1')

# Ver as primeiras linhas
df.head()
```

```python
df.columns
```

```python
df['SigTipoGeracao'].value_counts()
#Identificar quantas usinas por cada tipo de geração de energia
```

```python
import matplotlib.pyplot as plt

# Contar as usinas por tipo
contagem = df['SigTipoGeracao'].value_counts()

# Plotar o gráfico
plt.figure(figsize=(10,6))
contagem.plot(kind='bar')
plt.title('Quantidade de Usinas por Tipo de Geração')
plt.xlabel('Tipo de Geração')
plt.ylabel('Quantidade de Usinas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['SigUFPrincipal'].value_counts()
# Estados com mais usinas
```

```python
# Ver se a coluna é string, substituir virgula por ponto
df['MdaPotenciaFiscalizadaKw'] = df['MdaPotenciaFiscalizadaKw'].astype(str).str.replace(',', '.')

# Converter string para número
df['MdaPotenciaFiscalizadaKw'] = pd.to_numeric(df['MdaPotenciaFiscalizadaKw'], errors='coerce')

# Agrupar por tipo e somar a potência
potencia_por_tipo = df.groupby('SigTipoGeracao')['MdaPotenciaFiscalizadaKw'].sum().sort_values(ascending=False)

# Exibir resultado
potencia_por_tipo
```

```python
# Gráfico potência total por tipo de geração
plt.figure(figsize=(10,6))
potencia_por_tipo.plot(kind='bar')
plt.title('Potência Total por Tipo de Geração (em kW)')
plt.xlabel('Tipo de Geração')
plt.ylabel('Potência Total (kW)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Agrupar e somar a potência por estado (UF)
potencia_por_estado = df.groupby('SigUFPrincipal')['MdaPotenciaFiscalizadaKw'].sum().sort_values(ascending=False)

# Visualizar o ranking
potencia_por_estado
```

```python
# Gráfico de barras da potência por estado
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
potencia_por_estado.plot(kind='bar')
plt.title('Potência Instalada por Estado (em kW)')
plt.xlabel('Estado (UF)')
plt.ylabel('Potência Total (kW)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Categorizar os tipos de energia
renovaveis = ['UFV', 'UHE', 'EOL', 'PCH', 'CGH']

# Criar nova coluna com tipo de fonte
df['TipoFonte'] = df['SigTipoGeracao'].apply(lambda x: 'Renovável' if x in renovaveis else 'Não Renovável')
```

```python
# Agrupar por Estado e tipo de fonte, somando a potência
potencia_estado_fonte = df.groupby(['SigUFPrincipal', 'TipoFonte'])['MdaPotenciaFiscalizadaKw'].sum().unstack().fillna(0)
```

```python
# Gráfico de barras empilhadas
potencia_estado_fonte.sort_values(by='Renovável', ascending=False).plot(kind='bar', stacked=True, figsize=(14,6))
plt.title('Potência Total por Estado (Renovável vs Não Renovável)')
plt.xlabel('Estado (UF)')
plt.ylabel('Potência Total (kW)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Converter a coluna de data
df['DatEntradaOperacao'] = pd.to_datetime(df['DatEntradaOperacao'], errors='coerce')
```

```python
# Criar coluna com o ano de entrada em operação
df['AnoEntrada'] = df['DatEntradaOperacao'].dt.year
```

```python
# Contar quantidade de usinas por ano
usinas_por_ano = df['AnoEntrada'].value_counts().sort_index()
usinas_por_ano
```

```python
# Filtrar apenas anos válidos (acima de 2000 nesse caso)
usinas_por_ano_filtrado = usinas_por_ano[usinas_por_ano.index > 2000]
```

```python
plt.figure(figsize=(12,6))
usinas_por_ano_filtrado.plot(kind='bar')
plt.title('Usinas Entrando em Operação por Ano (a partir de 2001)')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Usinas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Contar quantidade de usinas por ano e tipo
usinas_ano_tipo = df.groupby(['AnoEntrada', 'SigTipoGeracao']).size().reset_index(name='Quantidade')
```

```python
# Agrupar
usinas_filtradas = usinas_ano_tipo[usinas_ano_tipo['AnoEntrada'] >= 2010]

# Pivotar os dados para deixar anos como índice e tipos como colunas
dados_pivot = usinas_filtradas.pivot(index='AnoEntrada', columns='SigTipoGeracao', values='Quantidade').fillna(0)

# Plotar com matplotlib
plt.figure(figsize=(14,6))
for coluna in dados_pivot.columns:
    plt.plot(dados_pivot.index, dados_pivot[coluna], marker='o', label=coluna)

plt.title('Evolução de Usinas por Tipo de Geração (a partir de 2010)')
plt.xlabel('Ano de Entrada em Operação')
plt.ylabel('Quantidade de Usinas')
plt.xticks(rotation=45)
plt.legend(title='Tipo de Geração')
plt.tight_layout()
plt.show()
```
